```python
import networkx as nx
import matplotlib.pyplot as plt

print(f'NetworkX version: {nx.__version__}')
```

```python
class Color:
    YELLOW  = '#e6db74'
    ORANGE  = '#fd971f'
    RED     = '#f92672'
    MAGENTA = '#fd5ff0'
    VIOLET  = '#ae81ff'
    BLUE    = '#66d9ef'
    CYAN    = '#a1efe4'
    GREEN   = '#a6e22e'
```

```python
G = nx.cycle_graph(24)
pos = nx.circular_layout(G)
nx.draw(G, pos, node_color=range(G.number_of_nodes()), node_size=800, cmap=plt.cm.cool)
plt.show()
```

```python
def print_and_draw(G):
    print(f'V {G.nodes}')
    print(f'E {G.edges}')
    plt.figure(figsize=(6, 6))
    nx.draw_networkx(G, node_size=800, node_color=Color.VIOLET)
    plt.axis('off')
    plt.show()

# G
G = nx.Graph()
G.add_node(3)                 # [3]
G.add_nodes_from(range(1, 6)) # [1,2,3,4,5]
print_and_draw(G)

# H
H = nx.path_graph(range(10, 16))
print_and_draw(H)

# G+H
G.add_nodes_from(H) # G.add_nodes_from(H.nodes)
print_and_draw(G)
```

```python
G.add_edge(1,2)
e = (2,3)
G.add_edge(*e)
G.add_edges_from([(14,15), (15,16), (16,17)])
G.add_edges_from(H.edges)
print_and_draw(G)
```

```python
G = nx.Graph()
G.add_nodes_from(range(1, 18))
G.add_edges_from([
    (1,2), (2,3), (3,5), (3,15), (4,6), (6,15), (6,7), (6,8), (9,13), (10,11), (11,12), (12,13), (13,14), (14,15), (15,16), (15,17)
])

print('|V|', G.number_of_nodes())
print('|E|', G.number_of_edges())
print('node_15\'s neighbors\t', list(G.neighbors(15)))
print('node_15\'s degree\t', G.degree(15))
print_and_draw(G)
```
